#Analyzing COVID Data Indian States and Union Territories


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```
df.head()
```

```
df = df.iloc[:-1,1:]
```

```
total_states = np.array(len(df['state_name']))
```

```
max(df['positive'])
```

##Total Positive Cases State Wise

```
plt.figure(figsize = (12,8))
sns.barplot(df['positive'],df['state_name'])
plt.xlim(1, max(df['positive'])+100)
plt.xlabel("Positive No. Of Cases")
plt.title("Total Positive Cases State Wise")
plt.show()
```

##Active No. Of Cases 

```
plt.figure(figsize = (12,8))
sns.barplot(df['new_active'],df['state_name'])
plt.xlim(1, max(df['new_active'])+100)
plt.xlabel("Active No. Of Cases")
plt.title("Total Active Cases State Wise")
plt.show()
```

##No. Of Deaths StateWise

```
plt.figure(figsize = (12,8))
sns.barplot(df['death'],df['state_name'])
plt.xlim(1, max(df['death'])+100)
plt.xlabel("No. Of deaths")
plt.title("Total deaths State Wise")
plt.show()
```
